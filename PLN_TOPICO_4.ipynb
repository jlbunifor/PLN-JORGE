# Criação de uma função que encontra o domínio de um email

```python
def captura_dominio(email):
    # divide o texto do e-mail pelo caractere @
    divisao_email=email.split('@')
    #separa o dominio do resto do email
    divisao_dominio=divisao_email[1].split('.')
    dominio=divisao_dominio[0]
    return dominio
```

```python
email_exemplo='npl_unifor@gmail.com'
print("O dominio do meu e-mail é:", captura_dominio(email_exemplo))
```

# Domínio de email

## Carregando bibliotecas necessárias

```python
#manupulação de dados em tabelas
import pandas as pd 
#plots de gráficos
import matplotlib.pyplot as plt
#manupalação de vetores
import numpy as np 
```

## Carregando dados

```python
tabela_email=pd.read_csv('emails.csv')
```

```python
#Algumas linhas da tabela
tabela_email.head()
```

## Capturar todos os domínios

```python
tabela_email['DOMINIO_EMAIL']=tabela_email['Email'].apply(captura_dominio)
```

```python
contagem_dominio=tabela_email['DOMINIO_EMAIL'].value_counts()
contagem_dominio
```

```python
plt.figure(figsize=(15,5),dpi=300)
plt.bar(contagem_dominio.index[0:30],contagem_dominio.values[0:30],color='orange',edgecolor='black')
plt.xticks(rotation=90,fontsize=20)
plt.grid(alpha=0.2)
plt.xlabel('Dominio',fontsize=20)
plt.ylabel('Contagem',fontsize=20)
plt.title('Top 30 - Domínios de e-mail mais frequentes')
plt.show()
```

```python
plt.figure(figsize=(15,5),dpi=300)

plt.scatter(np.log(range(1,len(contagem_dominio)+1)),np.log(contagem_dominio.values),color='orange',edgecolor='black')
plt.xticks(rotation=0,fontsize=20)
plt.grid(alpha=0.2)
plt.xlabel(r'$\ln(rank)$',fontsize=20)
plt.ylabel(r'$\ln(Contagem)$',fontsize=20)
plt.show()
```
