```python
# carregando o dado
df=pd.read_csv("CARROS_INSTAGRAM.csv")
df
```

# Embeddings 

```python
from sentence_transformers import SentenceTransformer
```

```python
model=SentenceTransformer('alfaneo/bertimbau-base-portuguese-sts')
```

```python
# Transformar textos em embeddings
texto_marca=list(df['TEXTO'].values)
matriz=model.encode(texto_marca)
```

```python
pd.DataFrame(matriz)
```

```python
# formato da matriz
matriz.shape
```

```python
from sklearn.manifold import TSNE
```

```python
X=matriz
tsne=TSNE(n_components=2,perplexity=10, n_iter=20000,init='random',random_state=42)
X_tse=tsne.fit_transform(X)
```

```python
plt.figure(dpi=300)
plt.title("REPRESENTAÇÃO 2D - TEXTOS",fontsize=15)
plt.scatter(X_tse[:,0],X_tse[:,1],edgecolor='black',lw=0.2)
plt.xticks([],[])
plt.yticks([],[])
```
